# 1. Load the dataset


```python
file_path = r"C:\Users\baliq\Downloads\MOCK_DATA (3).CSV"
#getting the number of rows
total_rows = sum(1 for _ in open(file_path)) - 1    # subtracting the header

#calculating 75% of the total rows\
rows_to_load = int(0.75 * total_rows)

#Reading only 75% of the data
df = pd.read_csv(file_path, nrows = rows_to_load)

df
```

# 2. Data Cleaning


```python
# Removing duplicates in the dataset
df_non_duplicate = df.drop_duplicates (keep = "first")
df_non_duplicate
```

There are no duplicates in the dataset

```python
# check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

There are no missing values

```python
df["Sign-up Date"] = pd.to_datetime (df["Sign-up Date"])   #Converting to proper date format
```

```python
df
```

```python
# replace all entries from the year 2020 with NaN.
df.loc[df["Sign-up Date"].dt.year == 2020, "Sign-up Date"] = np.NaN
df
```

```python
#Drop all rows where "Total Purchases" is less than 5 and customer is from Canada.
df= df.drop(df [((df["Total Purchases"]< 5) & (df["Country"] == "Canada"))].index)
df
```

There are only 2 Total purchase from Canada that is less than 5

## 3. Feature Engineering

```python
df["Sign-up Date"] = pd.to_datetime(df["Sign-up Date"])

# Calculate membership duration (in years)
current_year = pd.Timestamp.today().year
df["Membership Years"] = current_year - df["Sign-up Date"].dt.year

# Define function to calculate loyalty score
def assign_loyalty_score(row):
    if row["Total Purchases"] > 20:
        return 3
    elif 10 <= row["Total Purchases"] <= 20:
        return 2
    elif row["Total Purchases"] < 10:
        return 2 if row["Membership Years"] > 3 else 1

# Apply function to create new column
df["Loyalty Score"] = df.apply(assign_loyalty_score, axis=1)

print(df)
```

## 4. Data Aggregation & Filtering

```python
#Calculate avegrage and maximum purchase by country
Avg_purchases = df.groupby("Country")["Total Purchases"].mean().reset_index()
Max_purchases = df.groupby("Country")["Total Purchases"].max().reset_index()

print("Average Purchases per Country:")
print(Avg_purchases)

print("\nMax Purchases per Country:")
print(Max_purchases)
```

```python
# Excluding country mexico
excluded_country = "Mexico"
filtered_df = df[df["Country"] != "Mexico"]

# Find the most common loyalty score per country
most_common_loyalty = filtered_df.groupby("Country")["Loyalty Score"].agg(lambda x: x.mode().iloc[0])

# Display result
print(most_common_loyalty)
```

## 5. Data Export & Report Generation

```python
#Get today's date in the format DD-MM-YYYY
today_date = datetime.today().strftime("%d-%m-%Y")

# Generate filename in reverse order (YYYY_MM_DD)
filename = f"processed_data_{today_date.replace('-', '_')}.csv"

# Select the required columns
final_df = df[["Customer ID", "Loyalty Score", "Country", "Total Purchases"]]

# Shuffle the first 10 rows before displaying (without sorting)
shuffled_sample = final_df.sample(n=10, random_state=42)

# Save the DataFrame as a CSV
final_df.to_csv(filename, index=False)

# Print shuffled sample
print(shuffled_sample)
```
